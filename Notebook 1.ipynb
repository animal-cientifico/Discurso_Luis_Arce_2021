```
import re

# Leer el archivo de texto
with open('discurso.txt', 'r') as f:
    texto = f.read()

# Eliminar signos de puntuación y caracteres especiales
texto_limpio = re.sub(r'[^\w\s]', '', texto)

# Convertir la cadena de texto en una lista de palabras
palabras = texto_limpio.lower().split()

# Crear un diccionario para contar las palabras
contadores = {}

# Contar cada palabra en la lista
for palabra in palabras:
    if palabra in contadores:
        contadores[palabra] += 1
    else:
        contadores[palabra] = 1

# Ordenar el diccionario por frecuencia de palabras
contadores_ordenados = sorted(contadores.items(), key=lambda x: x[1], reverse=True)

# Imprimir las palabras y sus frecuencias
for palabra, frecuencia in contadores_ordenados:
    print(f'{palabra}: {frecuencia}')
```
